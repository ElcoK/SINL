```python
import os
import pandas as pd
import geopandas as gpd
```

```python
osm_files_path = os.path.join("C:\\Data",'OSM','feather')
damage_files_path =os.path.join("C:\\Data",'forecast','damage')
provinces = [x.split('.')[0] for x in os.listdir(osm_files_path)]
```

```python
collect_sum_ECMWF = {}
collect_sum_GFS =  {}

for province in provinces:
    try:
        # ECMWF
        model = 'ECMWF'
        
        collect_sums = {}
        
        for fc_time in ['039','041','043']:
            df = pd.read_feather(os.path.join(damage_files_path,'{}_{}_{}.ft'.format(province,model,fc_time)))
            collect_sums[fc_time] = df.damage_RM1L.sum(),df.damage_RM1L.sum()/df.m2.sum(),df.damage_RM1L.sum()/len(df)
            
            
        collect_sum_ECMWF[province] = pd.DataFrame.from_dict(collect_sums,orient='index').max(axis=0).values
        
        
        # GFS
        model = 'GFS'
        collect_sums = {}
        
        for fc_time in ['015','018']:
            df = pd.read_feather(os.path.join(damage_files_path,'{}_{}_{}.ft'.format(province,model,fc_time)))
            collect_sums[fc_time] = df.damage_RM1L.sum(),df.damage_RM1L.sum()/df.m2.sum(),df.damage_RM1L.sum()/len(df)
        
        
        collect_sum_GFS[province] = df.damage_RM1L.sum(),df.damage_RM1L.sum()/df.m2.sum(),df.damage_RM1L.sum()/len(df)        
        
        
    except:
        None
        
        
df_tot = pd.DataFrame.from_dict(collect_sum_ECMWF,orient='index',columns=['ECMWF_TOT','ECMWF_PER_M2','ECMWF_PER_BLD']
                      ).join(pd.DataFrame.from_dict(collect_sum_GFS,orient='index',columns=['GFS_TOT','GFS_PER_M2','GFS_PER_BLD'])
                            )        
```

```python
df_tot.to_excel('Eunice_impacts.xlsx')
```
